### Task 1: Load the dataset

```python
import pandas as pd

df = pd.read_csv("customers.csv")
df.head()
```

```python
df
```

### Task 2: Impute Nan with Median

```python
median_age = df["age"].median()
df["age"] = df["age"].fillna(median_age)
df
```

### Task 3: Encode gender as binary.

```python
df["gender"].unique()
```

```python
df["gender"] = df["gender"].apply(lambda x: 1 if x == "Male" else 0)
df
```

### Task 4: Encode city using one-hot encoding.

```python

# Using pd.get_dummies
# df = pd.get_dummies(df, columns=["city"], drop_first=False)
# df
```

```python
# Using sklearn onehot encoder
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
city_encoded = ohe.fit_transform(df[["city"]])
print(city_encoded)

df_city_encoded = pd.DataFrame(city_encoded, columns=ohe.get_feature_names_out(["city"]))
df_ohe = pd.concat([df, df_city_encoded], axis=1)
df_ohe
```

### Task 5: Normalize `salary` with **standard scaling**.

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

salary_normalised = ss.fit_transform(df_ohe[['salary']])
print(salary_normalised)

df_ohe['salary'] = salary_normalised
df_ohe
```

### Task 6: Wrap all these steps into a **scikit-learn Pipeline** for reusability.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

raw_df = pd.read_csv("customers.csv")
x_raw = raw_df.drop(columns=["purchased", "customer_id"])
y_raw = raw_df[["purchased"]]

y_raw


# define numerical and categorical transformers
numerical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(drop='first'))])


# combine into column transformer

preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer,['age', 'salary']), ('cat', categorical_transformer, ['gender', 'city'])])

# full pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

x_processed = pipeline.fit_transform(x_raw)
x_processed
```
